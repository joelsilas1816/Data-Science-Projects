```python
import string
def count_punc(sent):
    count=sum([1 for x in sent if x in string.punctuation])
    p = round(count/(len(sent)-sent.count(" ")) *100 ,2)
    return p
```

```python
df["punct%"] = df["tweet"].apply(lambda x : count_punc(x))
```

```python
df
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
import nltk # NAtural Language Toolkit
```

```python
nltk.download("stopwords")
```

```python
s_words = stopwords.words("english")
```

```python
sent = "Hello programmers. Data Science is a trending field! The future demands for skilled data Scientists. Want to know Why? The reason being predictive analysis"
```

```python
def clean_text(text):
    data = [x for x in text if x not in string.punctuation]
    # print(data)
    data = "".join(data)
    # print(data)
    data = [ps.stem(x) for x in data.split() if x not in s_words]
    return data
```

```python
clean_text(sent)
```

```python
# i/p
x = df.drop(["label","id"],axis=1)
# o/p
y = df["label"]
```

```python
x 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(analyzer=clean_text)
```

```python
tfidf.vocabulary_
```

```python
x_trans = tfidf.fit_transform(x["tweet"])
print(x_trans)
```

```python
x_trans.shape
```

```python
x_vect = pd.concat([x[["tweet_len","punct%"]].reset_index(drop=True),pd.DataFrame(x_trans.toarray())],axis=1)
```

```python
x_vect
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train, y_test = train_test_split(x_vect , y , stratify=y , random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
res = pd.DataFrame({"Actual" : y_test , "Predicted" : y_pred})
```

```python
res
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
print("Accuracy is :", accuracy_score(y_test,y_pred)*100,"%")
```
